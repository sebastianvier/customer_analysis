```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import kurtosis
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
data = pd.read_csv("data/marketing_customer_analysis.csv")
```

## 1) Preparing Data

```python
def standarize_column_names(data=data):
    data.columns = [i.lower().replace(" ", "_") for i in data.columns]
    
def split_X_and_y(target, data=data):
    X = data.drop([target], axis=1)
    y = data[[target]]
    return X,y

def split_numerical_and_categorical(X_data):
    X_numerical = X_data.select_dtypes(include=np.number)
    X_categorical = X_data.select_dtypes(exclude=np.number)
    return X_numerical, X_categorical
    
def preparing_data(target, data=data):
    """Returns based on targe a numerical x dataframe, a categorical x dataframe and the target as datafram"""
    standarize_column_names()
    X_data, y_data = split_X_and_y(target=target)
    X_numerical, X_categorical = split_numerical_and_categorical(X_data)
    return X_numerical, X_categorical, y_data

def days_to_numbers(df,column_name):
    df[column_name] = pd.to_datetime(df[column_name])
    df[column_name] = df[column_name].sub(pd.Timestamp(df[column_name].min()))
    df[column_name] = df[column_name].dt.days
    return df


def axes_generator(x,y):
    lst1 = [i for i in range(x)]
    lst2 = [i for i in range(y)]
    for first_num in lst1:
        for second_num in lst2:
            yield first_num, second_num
            
def check_outliers():
    fig, ax = plt.subplots(3,1, figsize=(5,10))
    for i in range(3):
        sns.boxplot(x=X_numerical.iloc[:,i], ax=ax[i], palette="summer")
    plt.tight_layout()
    plt.show
    
def get_outliers_indices(column_name, data = data):
    q1 = np.percentile(data[column_name], 25)
    q3 = np.percentile(data[column_name], 75) 
    iqr = q3-q1
    condition_1 = data[column_name] > q1 - 1.5* iqr
    condition_2 = data[column_name] < (q3 + 1.5 * iqr)
    
    ## You need to negate the interception of both conditions
    return (data[column_name][~(condition_1 & condition_2)].index)
```

```python
X_numerical, X_categorical, y_data = preparing_data("total_claim_amount")
```

```python
### Turns one column of the categorical values into numbers because they are dates.
X_categorical = days_to_numbers(X_categorical,"effective_to_date")
X_numerical["effective_to_date"] = X_categorical["effective_to_date"]
X_categorical = X_categorical.drop(["customer", "effective_to_date"],axis=1)
```

## 4. Linear Regression

```python
X = pd.concat([X_numerical, pd.DataFrame(encoded.todense())],axis=1)
lst = list(X.columns[:3])
X.columns = np.concatenate([lst,[str(i) for i in range(19)]])
X = scaler = StandardScaler().fit_transform(X)
y =y_data
```

### 4.1. Linear Regression with all data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

reg = LinearRegression().fit(X_train, y_train)
prediction = reg.predict(X_test)

mse = metrics.mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
error_percentage = rmse / y_test.mean()
r2 = metrics.r2_score(y_test,prediction)

print(f'Model rmse: {rmse}')
print(f'Error percentage: {error_percentage}')
print(f'Base model r2 score: {r2}')
```

### 4.2. Linear Regression dropping outliers

```python
######## Droping monthly premium auto outliers #######

X = pd.concat([X_numerical, pd.DataFrame(encoded.todense())],axis=1)

X = X.drop(mpa_outlier_indices, errors='ignore')
lst = list(X.columns[:3])
X.columns = np.concatenate([lst,[str(i) for i in range(19)]])
X = scaler = StandardScaler().fit_transform(X)
y =y_data
y = y_data.drop(mpa_outlier_indices, errors='ignore')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

reg = LinearRegression().fit(X_train, y_train)
prediction = reg.predict(X_test)

mse = metrics.mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
error_percentage = rmse / y_test.mean()
r2 = metrics.r2_score(y_test,prediction)

print(f'Model rmse: {rmse}')
print(f'Error percentage: {error_percentage}')
print(f'Base model r2 score: {r2}')
```

```python
######## Droping customer lifetime value outliers #######


X = pd.concat([X_numerical, pd.DataFrame(encoded.todense())],axis=1)

X = X.drop(clv_outlier_indices, errors='ignore')
lst = list(X.columns[:3])
X.columns = np.concatenate([lst,[str(i) for i in range(19)]])
X = scaler = StandardScaler().fit_transform(X)
y = y_data
y = y_data.drop(clv_outlier_indices, errors='ignore')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

reg = LinearRegression().fit(X_train, y_train)
prediction = reg.predict(X_test)

mse = metrics.mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
error_percentage = rmse / y_test.mean()
r2 = metrics.r2_score(y_test,prediction)

print(f'Model rmse: {rmse}')
print(f'Error percentage: {error_percentage}')
print(f'Base model r2 score: {r2}')
```

```python
######## Turning  monthly premium auto outliers to mean #######

X_mean = pd.concat([X_numerical, pd.DataFrame(encoded.todense())],axis=1)

mean = X_mean["monthly_premium_auto"].mean()
X_mean.at[mpa_outlier_indices, "monthly_premium_auto"] = mean
X_mean.columns = np.concatenate([lst,[str(i) for i in range(19)]])
X_mean = scaler = StandardScaler().fit_transform(X_mean)
y =y_data


X_train, X_test, y_train, y_test = train_test_split(X_mean, y, test_size=0.20, random_state=42)

reg = LinearRegression().fit(X_train, y_train)
prediction = reg.predict(X_test)

mse = metrics.mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
error_percentage = rmse / y_test.mean()
r2 = metrics.r2_score(y_test,prediction)

print(f'Model rmse: {rmse}')
print(f'Error percentage: {error_percentage}')
print(f'Base model r2 score: {r2}')
```

```python
######## Turning customer lifetime value outliers to mean #######

X_mean = pd.concat([X_numerical, pd.DataFrame(encoded.todense())],axis=1)
mean = X_mean["customer_lifetime_value"].mean()
X_mean.at[clv_outlier_indices, "customer_lifetime_value"] = mean
X_mean.columns = np.concatenate([lst,[str(i) for i in range(19)]])
X_mean = scaler = StandardScaler().fit_transform(X_mean)
y =y_data

X_train, X_test, y_train, y_test = train_test_split(X_mean, y, test_size=0.20, random_state=42)

reg = LinearRegression().fit(X_train, y_train)
prediction = reg.predict(X_test)

mse = metrics.mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
error_percentage = rmse / y_test.mean()
r2 = metrics.r2_score(y_test,prediction)

print(f'Model rmse: {rmse}')
print(f'Error percentage: {error_percentage}')
print(f'Base model r2 score: {r2}')
```

```python
######## Turning customer lifetime value outliers to mean #######

X_mean = pd.concat([X_numerical, pd.DataFrame(encoded.todense())],axis=1)
mean = X_mean["customer_lifetime_value"].mean()
X_mean.at[clv_outlier_indices, "customer_lifetime_value"] = mean
mean = X_mean["monthly_premium_auto"].mean()
X_mean.at[mpa_outlier_indices, "monthly_premium_auto"] = mean
X_mean.columns = np.concatenate([lst,[str(i) for i in range(19)]])
X_mean = scaler = StandardScaler().fit_transform(X_mean)
y =y_data

X_train, X_test, y_train, y_test = train_test_split(X_mean, y, test_size=0.20, random_state=42)

reg = LinearRegression().fit(X_train, y_train)
prediction = reg.predict(X_test)

mse = metrics.mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
error_percentage = rmse / y_test.mean()
r2 = metrics.r2_score(y_test,prediction)

print(f'Model rmse: {rmse}')
print(f'Error percentage: {error_percentage}')
print(f'Base model r2 score: {r2}')
```

For some reason the model doesn't improves a lot once the outliers are dropped. The best score is obtained when the customer lifetime value outliers are dropped. As most outliers are at the left side of the boxplot probably a better suggestion for a replacement of a value is a greater than the mean.
